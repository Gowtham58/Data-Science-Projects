```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

f=pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/Billionaire.csv")
f.head()
```

```python
f.isna().sum()
```

```python
f.dropna()
```

```python
f['NetWorth']=f['NetWorth'].str.strip('$')
f['NetWorth']=f['NetWorth'].str.strip('B')
f['NetWorth']=f['NetWorth'].astype(float)
```

```python
f['NetWorth']
```

```python
df=f.sort_values(by=['NetWorth'],ascending=False).head(10)
plt.figure(figsize=(20,10))
sns.histplot(x="Name",hue="NetWorth",data=df)
```

```python
a=f['Industry'].value_counts().head()
index=[i+"("+str(j)+")" for i,j in zip(a.index,a.values)]
industries=a.values
plt.figure(figsize=(5,5))
plt.pie(industries,labels=index)
central_circle=plt.Circle((0,0),0.5,color='white')
fig=plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font',size=12)
plt.title("Top 5 Industries with most number of Billionaires",size=20)
```
